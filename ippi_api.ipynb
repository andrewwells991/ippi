```python
import requests
import pandas as pd
from pandas import json_normalize 
```

```python
api_url = "https://ippi.mw/api/projects/query"
```

```python
response = requests.get(api_url)
```

```python
todo = {"start_date":"2021-01-01", "end_date":"2023-01-16"}
```

```python
response = requests.post(api_url, json = todo)
```

```python
d = response.json()
```

```python
df = json_normalize(d, 'projects')
df
```

Analysis of IPPI data

```python
# iterating the columns
for col in df.columns:
  print(col)
```

```python
#print the title column without header
for v in df['title']:
    print(v)
```
